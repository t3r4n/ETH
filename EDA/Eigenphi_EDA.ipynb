```python
# Imprimir valores únicos por columna (hasta un máximo de 20 valores por columna para no saturar)
for col in sandwich.columns:
    print(f"\n--- {col} ---")
    valores_unicos = sandwich[col].unique()
    if len(valores_unicos) > 20:
        print(valores_unicos[:20], "...")  # Mostrar solo los primeros 20
    else:
        print(valores_unicos)
```

## tienen implicito el simbolo porcentaje y pesos , no lo necesitamos


```python
exclude_cols = ['Date']

# Limpiar y convertir todas las demás columnas
for col in sandwich.columns:
    if col not in exclude_cols and sandwich[col].dtype == 'object':
        # Quitar $ , % y comas
        sandwich[col] = sandwich[col].str.replace(r'[\$,%,]', '', regex=True)
        # Convertir a numérico
        sandwich[col] = pd.to_numeric(sandwich[col], errors='coerce')

# Verificar tipos después de la conversión
sandwich.dtypes
```

```python
# Estadísticas descriptivas generales
pd.set_option('display.float_format', '{:,.2f}'.format)

summary = sandwich.describe().transpose()

# Calcular medias específicas de interés
roi_mean = sandwich['roi'].mean()
roi_median = sandwich['roi'].median()
profit_total = sandwich['ganancia_neta'].sum()

print(f"ROI promedio: {roi_mean:.2f}%")
print(f"Mediana ROI: {roi_median:.2f}%")
print(f"Ganancia total (Profit): {profit_total:.2f} ETH")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegurarse de que 'Date' sea tipo datetime
sandwich['Date'] = pd.to_datetime(sandwich['Date'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=sandwich, x='Date', y='ganancia_neta', color='mediumseagreen')
plt.title('Evolución de Ganancias Netas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ganancia Neta (ETH)')
plt.grid(True, alpha=0.2)
plt.show()
```

## muy poquitas en perdidas, entre enero y mayo hubo una extracción estrámbotica, lo atribuyo a tambien el avance en los validadores que hubo recientemente, pudo haber puesto tecnología disponible para extraer eso

```python
# Agrupar por año-mes y calcular la ganancia neta promedio
ganancia_mensual = sandwich.groupby(sandwich['Date'].dt.to_period('M'))['ganancia_neta'].mean().reset_index()
ganancia_mensual['Date'] = ganancia_mensual['Date'].dt.to_timestamp()  # Convertir Period a Timestamp para graficar

plt.figure(figsize=(14, 6))
sns.lineplot(data=ganancia_mensual, x='Date', y='ganancia_neta', marker='o', color='mediumseagreen')
plt.title('Ganancia Neta Promedio por Mes')
plt.xlabel('Mes')
plt.ylabel('Ganancia Neta Promedio (ETH)')
plt.grid(True, alpha=0.2)
plt.show()
```

Diciembre estuvo épico 

```python
# Agrupar por año y calcular la ganancia neta promedio
ganancia_anual = sandwich.groupby(sandwich['Date'].dt.year)['ganancia_neta'].mean().reset_index()
ganancia_anual.rename(columns={'Date': 'Año'}, inplace=True)  # Renombrar columna para claridad
ganancia_anual.rename(columns={ganancia_anual.columns[0]: 'Año'}, inplace=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=ganancia_anual, x=ganancia_anual.columns[0], y='ganancia_neta', marker='o', color='mediumseagreen', linewidth=2.5)
plt.title('Ganancia Neta Promedio por Año', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ganancia Neta Promedio (ETH)', fontsize=12)
plt.xticks(ganancia_anual[ganancia_anual.columns[0]])  # Mostrar todos los años
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar columnas numéricas (excepto 'Date')
numeric_cols = sandwich.select_dtypes(include=['float64', 'int64']).columns

# Definir dimensiones de la cuadrícula según número de variables
num_vars = len(numeric_cols)
cols = 2  # Ahora dos gráficas por fila
rows = (num_vars // cols) + int(num_vars % cols != 0)

# Crear figura
fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 4))  # Ajustar ancho por columna
axes = axes.flatten()

# Generar un boxplot por cada variable (vertical)
for i, col in enumerate(numeric_cols):
    sns.boxplot(y=sandwich[col], ax=axes[i], color='skyblue')
    axes[i].set_title(f'{col}', fontsize=12)
    axes[i].set_ylabel('')
    axes[i].set_xlabel(col)

# Eliminar ejes vacíos si sobran
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# Ajustar espaciado
plt.tight_layout()
plt.suptitle('Boxplots por Variable Numérica', fontsize=16, y=1.02)
plt.show()

```
