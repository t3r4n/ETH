# Revisemos qué tan limpios están los datos 


```python
# Imprimir valores únicos por columna (hasta un máximo de 20 valores por columna para no saturar)
for col in sandwich.columns:
    print(f"\n--- {col} ---")
    valores_unicos = sandwich[col].unique()
    if len(valores_unicos) > 20:
        print(valores_unicos[:20], "...")  # Mostrar solo los primeros 20
    else:
        print(valores_unicos)
```

## tienen implicito el simbolo porcentaje y pesos , no lo necesitamos


```python
exclude_cols = ['Date']

# Limpiar y convertir todas las demás columnas
for col in sandwich.columns:
    if col not in exclude_cols and sandwich[col].dtype == 'object':
        # Quitar $ , % y comas
        sandwich[col] = sandwich[col].str.replace(r'[\$,%,]', '', regex=True)
        # Convertir a numérico
        sandwich[col] = pd.to_numeric(sandwich[col], errors='coerce')

# Verificar tipos después de la conversión
sandwich.dtypes
```

```python
# Estadísticas descriptivas generales
pd.set_option('display.float_format', '{:,.2f}'.format)

summary = sandwich.describe().transpose()

# Calcular medias específicas de interés
roi_mean = sandwich['roi'].mean()
roi_median = sandwich['roi'].median()
profit_total = sandwich['ganancia_neta'].sum()

print(f"ROI promedio: {roi_mean:.2f}%")
print(f"Mediana ROI: {roi_median:.2f}%")
print(f"Ganancia total (Profit): {profit_total:.2f} ETH")

```
